## Arreglando deptos y agregando el codigo

```python
def formatingText(text):
    text = text.lower()
    text = re.sub('<.*?>', '', text)
    text = re.sub(':.*?:', '', text)
    text = re.sub(r"([^n\u0300-\u036f]|n(?!\u0303(?![\u0300-\u036f])))[\u0300-\u036f]+", r"\1", normalize( "NFD", text), 0, re.I)
    text = normalize( 'NFC', text)
    text = re.sub('[^a-z ]', '', text)
    return text
```

```python
crime_df['departamento']=crime_df.departamento.apply(formatingText)
```

```python
depto_code=pd.read_csv('codigos_departamentos.csv')
depto_code['name_dept_low']=depto_code['name_dept'].apply(formatingText)
depto_translate=depto_code.sort_values('name_dept').reset_index(drop=True)
depto_translate.head()
```

```python
cirme_dept=crime_df.sort_values('departamento').departamento.unique()
```

```python
depto_translate['crime_depto']=[difflib.get_close_matches(x,list(cirme_dept),n=1)[0] for x in depto_translate['name_dept_low']]
depto_translate
```

```python
depto_dict=depto_translate[['name_dept','crime_depto']].set_index('crime_depto').to_dict()
```

```python
crime_df['departamento']=crime_df['departamento'].replace(depto_dict['name_dept'])
```

```python
crime_df=crime_df.merge(depto_translate[['cod_dept','name_dept']],how='left',left_on='departamento',\
               right_on='name_dept')
```

```python
crime_df.departamento=crime_df.departamento.replace({'',np.nan})
```

## Tabla de deptos y municipios

```python
crime_df.crime.value_counts()
```

```python
int_crimes=['hurto_persona','lesiones_personales','violencia_intrafamiliar','delito_sexual','terrorismo','hurto_comercio','hurto_residencia','homicidio','amenaza']
int_crime_df=crime_df[crime_df['crime'].isin(int_crimes)]
int_crime_df.info()
```

```python
depto_long=int_crime_df.groupby(['cod_dept',int_crime_df.time.dt.year,'crime']).size().reset_index()
depto_long.rename(columns={0: "crime_count","time":'year'},inplace=True)
depto_long.cod_dept=depto_long.cod_dept.astype(int)
```

```python
dept_tot=crime_df.groupby(['cod_dept',crime_df.time.dt.year]).size().reset_index()
dept_tot.rename(columns={0:'total_crimes'},inplace=True)
dept_tot.cod_dept=dept_tot.cod_dept.astype(int)
```

```python
depto_wide=depto_long.pivot_table(index=['cod_dept','year'],columns='crime').reset_index()
depto_wide.columns=['cod_dept','year','amenaza','delito_sexual','homicidio','hurto_comercio','hurto_persona',\
                   'hurto_residencia','lesiones_personales','terrorismo','violencia_intrafamiliar']
depto_wide=depto_wide.merge(dept_tot,how='left',left_on=['cod_dept','year'],right_on=['cod_dept','time'])
depto_wide.drop(columns='time',inplace=True)
depto_wide.head()
```

```python
mun_long=int_crime_df.groupby(['cod_municipio',int_crime_df.time.dt.year,'crime']).size().reset_index()
mun_long.rename(columns={0: "crime_count","time":'year'},inplace=True)
mun_long.cod_municipio=mun_long.cod_municipio.astype(int)
```

```python
mun_long.head()
```

```python
mun_tot=crime_df.groupby(['cod_municipio',crime_df.time.dt.year]).size().reset_index()
mun_tot.rename(columns={0:'total_crimes'},inplace=True)
mun_tot.cod_municipio=mun_tot.cod_municipio.astype(int)
```

```python
mun_wide=mun_long.pivot_table(index=['cod_municipio','year'],columns='crime').reset_index()
mun_wide.columns=['cod_municipio','year','amenaza','delito_sexual','homicidio','hurto_comercio','hurto_persona',\
                   'hurto_residencia','lesiones_personales','terrorismo','violencia_intrafamiliar']
mun_wide=mun_wide.merge(mun_tot,how='left',left_on=['cod_municipio','year'],right_on=['cod_municipio','time'])
mun_wide.drop(columns='time',inplace=True)
mun_wide.head()
```
