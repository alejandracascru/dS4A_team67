```python
import pandas as pd
import requests
from io import StringIO
#!pip install pandasql
import pandasql as ps
import time
```

**Procesamiento de base de datos de persona**

```python
#sharing url from the csv file
orig_url='https://drive.google.com/file/d/1Jk2vPrB6GjHrrEFsdSUyd75EsCdSKaLb/view?usp=sharing'
file_id = orig_url.split('/')[-2]
dwn_url='https://drive.google.com/uc?export=download&id=' + file_id
url = requests.get(dwn_url).text
csv_raw = StringIO(url)
persona_acum = pd.read_csv(csv_raw,sep=";")
```

```python
start_time = time.time()
query = """select depto,depto_nombre,corte_anio
           ,round(sum(case when PER001 in (1,2) then 1 else 0 end * FEX),0) as PER001_T
           ,round(sum(case when PER001 = 1 then 1 else 0 end * FEX),0) as PER001_1
           ,round(sum(case when PER001 = 2 then 1 else 0 end * FEX),0) as PER001_2
           ,round(sum(case when PER002 = 1 then 1 else 0 end * FEX),0) as PER002_1
           ,round(sum(case when PER002 = 2 then 1 else 0 end * FEX),0) as PER002_2
           ,round(sum(case when PER002 = 3 then 1 else 0 end * FEX),0) as PER002_3
           ,round(sum(case when PER002 = 4 then 1 else 0 end * FEX),0) as PER002_4
           ,round(sum(case when PER002 = 5 then 1 else 0 end * FEX),0) as PER002_5
           ,round(sum(case when PER002 = 6 then 1 else 0 end * FEX),0) as PER002_6
           ,round(sum(case when PER004 = 1 then 1 else 0 end * FEX),0) as PER004_1
           ,round(sum(case when PER004 = 2 then 1 else 0 end * FEX),0) as PER004_2
           ,round(sum(case when PER009 = 1 then 1 else 0 end * FEX),0) as PER009_1
           ,round(sum(case when PER009 = 2 then 1 else 0 end * FEX),0) as PER009_2
           ,round(sum(case when PER009 = 3 then 1 else 0 end * FEX),0) as PER009_3
           ,round(sum(case when PER009 = 4 then 1 else 0 end * FEX),0) as PER009_4
           ,round(sum(case when PER011 = 1 then 1 else 0 end * FEX),0) as PER011_1
           ,round(sum(case when PER011 = 2 then 1 else 0 end * FEX),0) as PER011_2
           ,round(sum(case when PER013 = 1 then 1 else 0 end * FEX),0) as PER013_1
           ,round(sum(case when PER013 = 2 then 1 else 0 end * FEX),0) as PER013_2
           ,round(sum(case when PER013 = 3 then 1 else 0 end * FEX),0) as PER013_3
           ,round(sum(case when PER013 = 4 then 1 else 0 end * FEX),0) as PER013_4
           ,round(sum(case when PER014 = 1 then 1 else 0 end * FEX),0) as PER014_1
           ,round(sum(case when PER014 = 2 then 1 else 0 end * FEX),0) as PER014_2
           ,round(sum(case when PER014 = 3 then 1 else 0 end * FEX),0) as PER014_3
           ,round(sum(case when PER014 = 4 then 1 else 0 end * FEX),0) as PER014_4
           ,round(sum(case when PER014 = 5 then 1 else 0 end * FEX),0) as PER014_5
           ,round(sum(case when PER016 = 1 then 1 else 0 end * FEX),0) as PER016_1
           ,round(sum(case when PER016 = 2 then 1 else 0 end * FEX),0) as PER016_2
           ,round(sum(puntaje_sisben_3 * FEX),0) as sum_puntaje_sisben_3

           from persona_acum
           group by depto,depto_nombre,corte_anio"""
persona_agrup=ps.sqldf(query, locals())

persona_agrup['mean_puntaje_sisben_3']=round(persona_agrup['sum_puntaje_sisben_3']/persona_agrup['PER001_T'],2)
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
start_time = time.time()
query = """select depto,depto_nombre,corte_anio
           ,round(sum(case when HOG001 in (1,2,3,4) then 1 else 0 end * FEX),0) as HOG001_T
           ,round(sum(case when HOG001 = 1 then 1 else 0 end * FEX),0) as HOG001_1
           ,round(sum(case when HOG001 = 2 then 1 else 0 end * FEX),0) as HOG001_2
           ,round(sum(case when HOG001 = 3 then 1 else 0 end * FEX),0) as HOG001_3
           ,round(sum(case when HOG001 = 4 then 1 else 0 end * FEX),0) as HOG001_4
           ,round(sum(case when HOG004 = 1 then 1 else 0 end * FEX),0) as HOG004_1
           ,round(sum(case when HOG004 = 2 then 1 else 0 end * FEX),0) as HOG004_2
           ,round(sum(case when HOG004 = 3 then 1 else 0 end * FEX),0) as HOG004_3
           ,round(sum(case when HOG004 = 4 then 1 else 0 end * FEX),0) as HOG004_4
           ,round(sum(case when HOG009 = 1 then 1 else 0 end * FEX),0) as HOG009_1
           ,round(sum(case when HOG009 = 2 then 1 else 0 end * FEX),0) as HOG009_2
           ,round(sum(case when HOG009 = 3 then 1 else 0 end * FEX),0) as HOG009_3
           ,round(sum(case when HOG009 = 4 then 1 else 0 end * FEX),0) as HOG009_4
           ,round(sum(case when HOG009 = 5 then 1 else 0 end * FEX),0) as HOG009_5
           ,round(sum(case when HOG009 = 6 then 1 else 0 end * FEX),0) as HOG009_6
           ,round(sum(case when HOG009 = 7 then 1 else 0 end * FEX),0) as HOG009_7
           ,round(sum(case when HOG009 = 8 then 1 else 0 end * FEX),0) as HOG009_8
           ,round(sum(case when HOG009 = 9 then 1 else 0 end * FEX),0) as HOG009_9
           ,round(sum(case when HOG016 = 1 then 1 else 0 end * FEX),0) as HOG016_1
           ,round(sum(case when HOG016 = 2 then 1 else 0 end * FEX),0) as HOG016_2
           ,round(sum(case when HOG016 = 3 then 1 else 0 end * FEX),0) as HOG016_3
           ,round(sum(case when HOG016 = 4 then 1 else 0 end * FEX),0) as HOG016_4
           ,round(sum(case when HOG016 = 5 then 1 else 0 end * FEX),0) as HOG016_5
           ,round(sum(case when HOG016 = 6 then 1 else 0 end * FEX),0) as HOG016_6
           ,round(sum(case when HOG017 = 1 then 1 else 0 end * FEX),0) as HOG017_1
           ,round(sum(case when HOG017 = 2 then 1 else 0 end * FEX),0) as HOG017_2
           ,round(sum(case when HOG017 = 3 then 1 else 0 end * FEX),0) as HOG017_3
           ,round(sum(case when HOG017 = 4 then 1 else 0 end * FEX),0) as HOG017_4
           ,round(sum(case when HOG021 = 1 then 1 else 0 end * FEX),0) as HOG021_1
           ,round(sum(case when HOG021 = 2 then 1 else 0 end * FEX),0) as HOG021_2
           ,round(sum(case when HOG022 = 1 then 1 else 0 end * FEX),0) as HOG022_1
           ,round(sum(case when HOG022 = 2 then 1 else 0 end * FEX),0) as HOG022_2
           ,round(sum(case when HOG026 = 1 then 1 else 0 end * FEX),0) as HOG026_1
           ,round(sum(case when HOG026 = 2 then 1 else 0 end * FEX),0) as HOG026_2
           ,round(sum(HOG032 * FEX),0) as HOG032
           from hogar_acum
           group by depto,depto_nombre,corte_anio"""
hogar_agrup=ps.sqldf(query, locals())

print("--- %s seconds ---" % (time.time() - start_time))
```

**Procesamiento de base de datos de vivienda**

```python
#sharing url from the csv file
orig_url='https://drive.google.com/file/d/1qiYLZnKuESI3WA5EBzwwd_CgiS5tL4Cg/view?usp=sharing'
file_id = orig_url.split('/')[-2]
dwn_url='https://drive.google.com/uc?export=download&id=' + file_id
url = requests.get(dwn_url).text
csv_raw = StringIO(url)
vivienda_acum = pd.read_csv(csv_raw,sep=";")
```

```python
start_time = time.time()
query = """select depto,depto_nombre,corte_anio
           ,round(sum(case when VIV001 in (1,2,3,4) then 1 else 0 end * FEX),0) as VIV001_T
           ,round(sum(case when VIV001 = 1 then 1 else 0 end * FEX),0) as VIV001_1
           ,round(sum(case when VIV001 = 2 then 1 else 0 end * FEX),0) as VIV001_2
           ,round(sum(case when VIV001 = 3 then 1 else 0 end * FEX),0) as VIV001_3
           ,round(sum(case when VIV001 = 4 then 1 else 0 end * FEX),0) as VIV001_4
           ,round(sum(case when VIV004 = 1 then 1 else 0 end * FEX),0) as VIV004_1
           ,round(sum(case when VIV004 = 2 then 1 else 0 end * FEX),0) as VIV004_2
           ,round(sum(case when VIV005 = 1 then 1 else 0 end * FEX),0) as VIV005_1
           ,round(sum(case when VIV005 = 2 then 1 else 0 end * FEX),0) as VIV005_2
           ,round(sum(case when VIV007 = 1 then 1 else 0 end * FEX),0) as VIV007_1
           ,round(sum(case when VIV007 = 2 then 1 else 0 end * FEX),0) as VIV007_2
           ,round(sum(case when VIV009 = 1 then 1 else 0 end * FEX),0) as VIV009_1
           ,round(sum(case when VIV009 = 2 then 1 else 0 end * FEX),0) as VIV009_2
           from vivienda_acum
           group by depto,depto_nombre,corte_anio"""
vivienda_agrup=ps.sqldf(query, locals())

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
info_departamentos = pd.merge(persona_agrup, hogar_agrup, how='left', on=['depto', 'depto_nombre','corte_anio'])
info_departamentos = pd.merge(info_departamentos, vivienda_agrup, how='left', on=['depto', 'depto_nombre','corte_anio'])
```

```python
from google.colab import files
info_departamentos.to_csv('info_nivel_departamentos.csv') 
files.download('info_nivel_departamentos.csv')
```
